```python
def init_browser():
    executable_path = {'executable_path': 'chromedriver.exe'}
    browser = Browser('chrome', **executable_path, headless=False)
    return browser

browser = init_browser()
```

```python
browser.visit(test_url)
```

```python
html = browser.html
# Parse with soup
soup = bs(html, 'html.parser')
```

```python
print(soup.prettify())
```

```python
table = soup.find("table")

# The first tr contains the field names.
headings = [th.get_text().strip() for th in table.find("tr").find_all("th")]

print(headings)

datasets = []
for row in table.find_all("tr")[1:]:
    dataset = dict(zip(headings, (td.get_text() for td in row.find_all("td"))))
    datasets.append(dataset)

print(datasets)
```

```python
temp = soup.findAll('div',attrs={"class":"table_outer_container"})
```

```python
# four factor table
temp[1].findAll('th', attrs={"class":"left "})[0].find("a").text
temp[1].findAll('th', attrs={"class":"left "})[1].find("a").text
temp[1].findAll('th', attrs={"class":"left "})[0].find("a").text == 'LAL'
```
